```python
import jieba
import jieba.posseg as pseg
import jieba.analyse


article = []
jieba.load_userdict('./argiculture/key.txt')
for i in range(421):
    text = str(data[i][0])
    a = list(jieba.cut(text, cut_all=False, HMM=True))

    s = ''
    for term in a:
        if '\u4e00' <= term[0] <= '\u9fff':
            s += (term+' ') 
    article.append(s)
print(article)


    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
import csv

vectorizer = CountVectorizer()
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(vectorizer.fit_transform(article))
word = vectorizer.get_feature_names()
weight = tfidf.toarray()

print(len(word))
total = 0


key = list(open('./argiculture/key.txt','r',encoding = 'utf-8'))
keyword = []
for i in key:
    x= i.split('\t')
    for t in x :
        if t != '' and t != '\n' and t != ' ':
            keyword.append(t)
def symmetric(t1,t2):
    text = (t1)
    a = list(jieba.cut(text, cut_all=False, HMM=True))
    text1 = (t2)
    b = list(jieba.cut(text1, cut_all=False, HMM=True))
    s1 = set()
    
    for term in a:
        if term in keyword:     
            s1.add(term)
    s2 = set()
    for term in b:
        if term in keyword:
            s2.add(term)
    s1 = list(s1)
    s2 = list(s2)
    for i in s1:
        for j in r:
            if i in j and i != j[0] and i in s1:
                s1.remove(i)
                s1.append(j[0])
                #print(s1)
    for i in s2:
        for j in r:
            if i in j and i != j[0] and i in s2:
                s2.remove(i)
                s2.append(j[0])
                #print(s2)
    x = len(list(set(s1)-set(s2)))
    y = len(list(set(s2)-set(s1)))
    if x-y <= 1:
        if y-x <=1:
            return 0
        else:
            return 1
    else:
        if y-x >=2:
            return 0
        else:
            return 2
    
used = np.zeros((421,421))
file = open("submission_example.csv",'w',encoding = 'utf-8',newline = '')
f = csv.writer(file)
f.writerow(['Test', 'Reference'])
for i in range(421):
    for j in range(421):
        #print('--------------------',i,j)
        if sum(weight[i]*weight[j]) > 0.45 and i < j:
            #print(data[i][1],data[j][1])
            p = symmetric(str(data[i][0]),str(data[j][0]))
            if p == 0:
                used[i][j] = used[j][i] = 1
            if p == 1:
                used[i][j] = 1; used[j][i] = 0
            if p == 2:
                used[i][j] = 0; used[j][i] = 1
                
        if used[i][j] == 1:
            total +=1
            f.writerow([data[i][1],data[j][1]])
print(total)

file.close()

```
